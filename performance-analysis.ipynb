```python
import os
import pandas as pd

df = pd.read_excel(os.path.join('data', 'PREVI-OPTIONS.xls'))
df = df[['Code ISIN', 'Unité de compte', 'Date création du fonds', 'Catégorie MorningStar', 'Performances annuelles', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Performances glissantes', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'Unnamed: 26', 'Volatilités', 'Unnamed: 34', 'Unnamed: 35', 'Ratio de Sharpe', 'Frais de gestion']]
df = df.rename(columns={'Code ISIN': 'ISIN', 'Unité de compte': 'Unit of account', 'Date création du fonds': 'Fund creation date', 'Catégorie MorningStar': 'MorningStar category', 'Performances annuelles': '2018', 'Unnamed: 16': '2019', 'Unnamed: 17': '2020', 'Unnamed: 18': '2021', 'Unnamed: 19': '2022', 'Unnamed: 20': '2023', 'Unnamed: 21': '2024', 'Performances glissantes': '10 years', 'Unnamed: 23': '5 years', 'Unnamed: 24': '3 years', 'Unnamed: 25': '1 year', 'Unnamed: 26': 'YTD', 'Volatilités': 'Volatility 5 years', 'Unnamed: 34': 'Volatility 3 years', 'Unnamed: 35': 'Volatility 1 year','Frais de gestion':'Fees'})
df = df.iloc[1:]
df = df.fillna(0)
df['Fees'] = df['Fees'].str.replace('%', '').str.replace(',', '.').astype(float)/100
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Group by 'MorningStar category' and calculate the average of '5 years'
average_5_years = df.groupby('MorningStar category')['5 years'].mean() / 5 * 100
top_average_5_years = average_5_years.sort_values(ascending=False).head(8)

# Plotting the top 8
fig, ax = plt.subplots(figsize=(12, 7))
bars = ax.bar(
    x=np.arange(top_average_5_years.size),
    height=top_average_5_years,
    tick_label=top_average_5_years.index.str.wrap(18),
    color="cornflowerblue",
)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')
ax.tick_params(bottom=False, left=False, labelsize=9, pad=12)
ax.set_axisbelow(True)
ax.yaxis.grid(True, color='#EEEEEE')
ax.xaxis.grid(False)
for bar in bars:
  ax.text(
      bar.get_x() + bar.get_width() / 2,
      bar.get_height() + 0.3,
      round(bar.get_height(), 1),
      horizontalalignment='center',
      color="cornflowerblue",
      weight='bold'
  )
ax.set_xlabel('MorningStar Category', labelpad=15, color='#333333', fontsize=11)
ax.set_ylabel('Average Performance (%)', labelpad=15, color='#333333', fontsize=11)
ax.set_title('Average Performance by MorningStar Category', pad=15, color='#333333', weight='bold', fontsize=13)
plt.tight_layout()
plt.savefig(os.path.join('visualizations', 'average_performance_category.png'))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create boolean mask for missing data across years
df_years = df[['10 years', '5 years', '3 years', '1 year']]

# Reshape data for visualization
df_years_melt = df_years.melt(var_name='Year', value_name='Has Data')
df_years_missing = df_years_melt[df_years_melt['Has Data'] == 0]  # Use == 0 instead of ~ operator

# Create figure and style settings
plt.figure(figsize=(4, 6))

# Plot missing data distribution
ax = sns.countplot(
    data=df_years_missing,
    x='Year',
    color='#FF6B6B',
    alpha=0.8
)

# Customize plot appearance
plt.title('Missing Data Distribution by Year', 
          fontsize=14, 
          pad=15,
          fontweight='bold')
plt.xlabel('Time Period', fontsize=12, labelpad=10)
plt.ylabel('Number of Missing Values', fontsize=12, labelpad=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add value labels on top of bars
for i in ax.containers:
    ax.bar_label(i, padding=3)

# Adjust layout and display
plt.tight_layout()
plt.savefig(os.path.join('visualizations', 'missing_data_distribution.png'))
```

```python
import numpy as np
import seaborn as sns

# Set style and color palette
sns.set_palette("husl")

# Define time periods and corresponding volatility measures
years = ['1 year', '3 years', '5 years']
volatility_years = ['Volatility 1 year', 'Volatility 3 years', 'Volatility 5 years']

# Plot performance vs volatility for each time period
for i, (year, volatility_year) in enumerate(zip(years, volatility_years)):
    # Create figure for this time period
    plt.figure(figsize=(16, 7), dpi=200)
    
    # Get top 15 performers for this period
    df_positive_performance = df[df[year] >= 0].sort_values(by=year, ascending=False).head(15)
    
    # Create custom colormap
    colors = sns.color_palette("husl", n_colors=len(df_positive_performance))
    
    # Plot each fund
    for j, (index, row) in enumerate(df_positive_performance.iterrows()):
        # Calculate annualized performance
        years_in_period = int(year.split()[0])
        annualized_performance = (1 + row[year]) ** (1 / years_in_period) - 1
        performance_value = annualized_performance * 100
        volatility_value = row[volatility_year] * 100
        
        # Add data point with improved styling
        plt.scatter(performance_value, volatility_value, 
                   color=colors[j], s=100, alpha=0.7,
                   label=row["Unit of account"])
        
        # Add labels with improved positioning and style
        plt.annotate(row["Unit of account"], 
                    (performance_value, volatility_value),
                    textcoords="offset points", 
                    xytext=(0, 10),
                    ha='center',
                    fontsize=8,
                    bbox=dict(facecolor='white', edgecolor='none', alpha=0.7))
    
    # Customize plot appearance
    plt.title(f'Performance vs Volatility - {year}', 
              fontsize=14, pad=20, fontweight='bold')
    plt.xlabel('Annualized Performance (%)', fontsize=12, labelpad=10)
    plt.ylabel('Volatility (%)', fontsize=12, labelpad=10)
    
    # Enhance grid and spines
    plt.grid(True, linestyle='--', alpha=0.7)
    for spine in plt.gca().spines.values():
        spine.set_linewidth(0.5)
    
    # Adjust legend
    plt.legend(bbox_to_anchor=(1.05, 1), 
              loc='upper left',
              borderaxespad=0.,
              fontsize=8)

    plt.tight_layout()
    plt.savefig(f'visualizations/performance_volatility_analysis_{year.replace(" ", "_")}.png', 
                bbox_inches='tight', dpi=300)
```

```python
# Plotting the Sharpe ratio against 1, 3, 5 years performance annualized
fig, axs = plt.subplots(3, 1, figsize=(18, 22), dpi=100)

# Define style parameters
TITLE_SIZE = 16
LABEL_SIZE = 12
ANNOTATION_SIZE = 10
MARKER_SIZE = 100
ALPHA = 0.7

years = ['1 year', '3 years', '5 years']
for i, year in enumerate(years):
    # Keep only the 12 best performing funds for each year
    df_positive_performance = df[df[year] >= 0].sort_values(by=year, ascending=False).head(12)
    
    # Create custom colormap
    colors = sns.color_palette("husl", n_colors=len(df_positive_performance))

    for j, (index, row) in enumerate(df_positive_performance.iterrows()):
        # Calculate annualized performance
        years_in_period = int(year.split()[0])
        annualized_performance = (1 + row[year]) ** (1 / years_in_period) - 1
        performance_value = annualized_performance * 100
        sharpe_ratio = row['Ratio de Sharpe']
        
        # Add data point with improved styling
        axs[i].scatter(sharpe_ratio, performance_value, 
                      color=colors[j], 
                      s=MARKER_SIZE, 
                      alpha=ALPHA,
                      label=row["Unit of account"])
        
        # Add labels with improved positioning and style
        axs[i].annotate(row["Unit of account"], 
                       (sharpe_ratio, performance_value),
                       textcoords="offset points", 
                       xytext=(0, 10),
                       ha='center',
                       fontsize=ANNOTATION_SIZE,
                       bbox=dict(facecolor='white', 
                                edgecolor='none', 
                                alpha=ALPHA))

    # Customize subplot appearance
    axs[i].set_title(f'Sharpe Ratio vs {year} Performance', 
                     fontsize=TITLE_SIZE, 
                     pad=20, 
                     fontweight='bold')
    axs[i].set_xlabel('Sharpe Ratio', fontsize=LABEL_SIZE, labelpad=10)
    axs[i].set_ylabel('Annualized Performance (%)', fontsize=LABEL_SIZE, labelpad=10)
    
    # Enhance grid and spines
    axs[i].grid(True, linestyle='--', alpha=0.7)
    for spine in axs[i].spines.values():
        spine.set_linewidth(0.5)
        
    # Adjust legend
    axs[i].legend(bbox_to_anchor=(1.05, 1), 
                 loc='upper left',
                 borderaxespad=0.,
                 fontsize=ANNOTATION_SIZE)

plt.tight_layout()
plt.savefig('visualizations/sharpe_ratio_performance.png', bbox_inches='tight', dpi=300)
```
